# QR code Generator

```python
#programming_fever
import qrcode
import matplotlib.pyplot as plt
#data
data = "https://www.instagram.com/programming_fever/"
# output file name
filename = "programming_fever.png"
# generate qr code
img = qrcode.make(data)
# save img to a file
img.save(filename)
plt.imshow(img,cmap='gray')
```
